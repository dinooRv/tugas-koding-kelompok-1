Matrix Multiplication

soal 1

using Numpy

```python
import numpy as np
```

```python
#buat matrix A dengan ukuran 2x3 dan vector b berukuran 3x1 menggunakan numpy
A = np.array([[1, 2, 3], [4, 5, 6]])
A
```

```python
b = np.array([[1],
              [2],
              [3]])
b
```

using pytorch

```python
import torch
```

```python
#buat matrix A_pt dengan ukuran 2x3 dan vector b_pt berukuran 3x1 menggunakan pytorch
A_pt = torch.tensor([[1, 2, 3], [4, 5, 6]])
A_pt
```

```python
b_pt = torch.tensor([[1],
                      [2],
                      [3]])
b_pt
```

using tensorflow

```python
import tensorflow as tf
```

```python
#buat matrix A_tf dengan ukuran 2x3 dan vector b_tf berukuran 3x1 menggunakan tensorflow
A_tf = tf.Variable([[1, 2, 3], [4, 5, 6]])
A_tf
```

```python
b_tf = tf.Variable([[1],
                    [2],
                    [3]])
b_tf
```

soal 3

```python
#kalikan matrix A dengan vector b
np.dot(A, b) # even though technically dot products are between vectors only
```

```python
#kalikan matrix A_pt dengan vector b_pt
hasil_pt = torch.matmul(A_pt, b_pt)
hasil_pt
```

```python
#kalikan matrix A_tf dengan vector b_tf
hasil_tf = tf.matmul(A_tf, b_tf)
hasil_tf
```

```python
#buat matrix B_tf dengan ukuran 3x3 menggunakan tensorflow
B_tf = tf.Variable([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
B_tf
```
